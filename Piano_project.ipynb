```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

**Mini Piano Synthesizer prototype**

```
import numpy as np
from IPython.display import Audio, display
import io
import base64

# ----- Configuration -----
SAMPLE_RATE = 44100  # Hz
DURATION = 0.5       # seconds per note

# Map piano keys (white keys C–B) to frequencies (Hz)
NOTES = {
    'C': 261.63,
    'D': 293.66,
    'E': 329.63,
    'F': 349.23,
    'G': 392.00,
    'A': 440.00,
    'B': 493.88
}

# ----- Audio synthesis -----
def play_tone(frequency, duration=DURATION):
    t = np.linspace(0, duration, int(SAMPLE_RATE * duration), False)
    # Generate a sine wave at the given frequency
    wave = np.sin(frequency * t * 2 * np.pi)
    # Normalize to 16-bit range
    audio = (wave * 32767).astype(np.int16)
    # Create and display the audio
    display(Audio(audio, rate=SAMPLE_RATE))

# Example usage:
# Play notes A, B, C, D, E, F, G
play_tone(NOTES['A'])
play_tone(NOTES['B'])
play_tone(NOTES['C'])
play_tone(NOTES['D'])
play_tone(NOTES['E'])
play_tone(NOTES['F'])
play_tone(NOTES['G'])
```
